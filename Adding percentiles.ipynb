```python
i = 5 #so starts at region
while i < len(df.columns):
    if is_numeric_dtype(df.iloc[:, i]):
        new_column = pd.qcut(df.iloc[:, i], np.linspace(0,1,11), duplicates = 'drop') #potentially add labels=np.linspace(0.1,1,10), 
        #now that percentile column is made, insert column next to relevant orignal column
        df.insert(i+1, 
                  str(df.columns[i]) + ' percentile', 
                 new_column, allow_duplicates = False)
        i=i+2
    else: 
        i=i+1
```
